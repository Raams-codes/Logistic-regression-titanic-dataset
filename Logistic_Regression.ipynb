### Data separation

```python
x=df.drop(["Survived"],axis=1)
y=df["Survived"]
```

```python
x.head()
```

```python
y.head()
```

#### Train test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

### Min Max Scaler

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
col=x_train.columns.to_list()
col
```

```python
x_train
```

```python
scaler.fit(x_train)
```

```python
scaler.fit(x_test)
```

```python
x_train_scaled=scaler.transform(x_train)
x_train_scaled=pd.DataFrame(x_train_scaled,columns=col)
x_train_scaled
```

```python
x_test_scaled=pd.DataFrame((scaler.transform(x_test)),columns=col)
x_test_scaled
```

### Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
# Building Model
LogReg=LogisticRegression()
```

```python
LogReg.fit(x_train_scaled,y_train)
```

### Prediction

```python
pred=LogReg.predict(x_test_scaled)
```

```python
train_pred=LogReg.predict(x_train_scaled)
```

```python
train_f1score=f1_score(train_pred,y_train)
train_f1score
```

```python
f1score=f1_score(pred,y_test)
```

```python
f1score
```

```python
accuracy=accuracy_score(pred,y_test)
accuracy
```

```python
conf_matrix=confusion_matrix(pred,y_test)
conf_matrix
```

```python
report=classification_report(pred,y_test)
print(f"classification report :\n {report}")
```

### Prob prediction

```python
prob_pred=LogReg.predict_proba(x_test_scaled)
prob_pred=prob_pred[:,1]
```

```python
prob_predicted=[]
for i in range(0,len(prob_pred)):
    if ((prob_pred[i])>=0.55):
        prob_predicted.append(1)
    else:
        prob_predicted.append(0)
```

```python
prob_matrix=classification_report(prob_predicted,y_test)
print(f"Classification Report of prob_prediction: \n\n{prob_matrix}")
```

### Visualizing Coefficiants

```python
co_ef=(LogReg.coef_)
co_ef=(co_ef).reshape(-1)
co_ef
```

```python
coef_df=pd.DataFrame({"Co_ef":co_ef},index=(x.columns))
coef_df
```

```python
plt.bar(coef_df.index,coef_df["Co_ef"])
plt.xticks(rotation=90)
plt.show()
```
